#### Problema 1:
#### Desarrolle una funci´on la cual reciba dos par´ametros, una imagen y un entero llamado color, lafunci´on debe devolver una imagen la cual tenga activos los canales de color segu´n los siguientes puntos:
#### Si el par´ametro color vale 1, la imagen debe mostrar activos u´nicamente el color azul.
#### Si el par´ametro color vale 2, la imagen debe mostrar activos u´nicamente el color verde.
#### Si el par´ametro color vale 3, la imagen debe mostrar activos u´nicamente el color rojo.
#### Si el par´ametro color vale 10, la imagen debe mostrar activos u´nicamente los colores rojo y verde.
#### Si el par´ametro color vale 20, la imagen debe mostrar activos u´nicamente los colores verde y azul.
#### Si el par´ametro color vale 30, la imagen debe mostrar activos u´nicamente los colores azul yrojo.

```python
def color_imagen(ruta_imagen, color):
    # Leer la imagen usando OpenCV
    img = cv2.imread(ruta_imagen)
    
    # Verificar si la imagen se cargó correctamente
    if img is None:
        print("No se pudo cargar la imagen. Por favor, verifica la ruta.")
        return None

    # Convertir la imagen a RGB ya que OpenCV la carga en BGR por defecto
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Inicializar una matriz de ceros del mismo tamaño que la imagen
    zeros = np.zeros(img_rgb.shape[:2], dtype="uint8")

    # Crear una copia de la imagen en escala de grises
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

    if color == 1:   # Canal Azul
        img_modificada = cv2.merge([img_gray, zeros, zeros])
    elif color == 2:  # Canal Verde
        img_modificada = cv2.merge([zeros, img_gray, zeros])
    elif color == 3:  # Canal Rojo
        img_modificada = cv2.merge([zeros, zeros, img_gray])
    elif color == 20:  # Canal Verde y Azul
        img_modificada = cv2.merge([zeros, img_gray, img_gray])
    else:
        print("El valor de color no es válido. Debe ser 1 (azul), 2 (verde), 3 (rojo) o 20 (verde y azul).")
        return None

    return img_modificada
```

```python
# Cargar la imagen desde la ruta
imagen = cv2.imread('C:\\Users\\crist\\Lab1\\Panda.jpg')

# Crear una figura de Matplotlib
fig = plt.figure(figsize=(10, 5))

# Agregar un subplot a la figura
ax = fig.add_subplot(1, 2, 1)

# Mostrar la imagen convertida a RGB en el subplot
ax.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

# Establecer el título del subplot
ax.set_title('Imagen Original')

# Desactivar los ejes del gráfico
ax.axis('off')

# Mostrar la figura
plt.show()
```

```python
img_modificada = color_imagen('C:\\Users\\crist\\Lab1\\panda.jpg', 20)

# Crear una figura y un subplot para mostrar la imagen
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 2, 2)

# Mostrar la imagen modificada en el subplot
ax.imshow(img_modificada)

# Establecer el título del subplot
ax.set_title('Imagen con filtro')

# Activar los ejes del subplot
ax.axis('on')

# Mostrar la figura con la imagen
plt.show()
```

#### Problema 3:
#### Cree una funci´on que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su funci´on debe crear 3 im´agenes como salida. Para entregar este ejercicio debe incluir una las im´agenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2

```python
def divide_image_into_grayscale(image_path):
    # Cargando la imagen
    image = cv2.imread(image_path)
    
    # Verificar si la imagen se cargó correctamente
    if image is None:
        print("Error al cargar la imagen.")
        return

    # Separar la imagen en canales
    b_channel, g_channel, r_channel = cv2.split(image)

    # Crear una figura en Matplotlib para mostrar los canales de color
    fig, axs = plt.subplots(1, 3, figsize=(20, 20))

    # Mostrar el canal azul
    axs[0].imshow(b_channel, cmap='gray')
    axs[0].set_title('Canal Azul')
    axs[0].axis('off')

    # Mostrar el canal verde
    axs[1].imshow(g_channel, cmap='gray')
    axs[1].set_title('Canal Verde')
    axs[1].axis('off')

    # Mostrar el canal rojo
    axs[2].imshow(r_channel, cmap='gray')
    axs[2].set_title('Canal Rojo')
    axs[2].axis('off')

    # Mostrar la figura
    plt.show()

    # Guardar las imágenes de salida
    cv2.imwrite('blue_channel.jpg', b_channel)
    cv2.imwrite('green_channel.jpg', g_channel)
    cv2.imwrite('red_channel.jpg', r_channel)

# Uso de la función con la imagen del panda
divide_image_into_grayscale('C:\\Users\\crist\\Lab1\\Panda.jpg')
```

#### Problema 4:
#### Cree una funci´on que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritm´etico para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una l´ınea vertical la cual muestre el valor de la media de la distribuci´on

```python
def show_histogram(image_path):
    # Cargar la imagen
    image = cv2.imread(image_path)
    
    # Comprobar si la imagen se cargó correctamente
    if image is None:
        print("Error al cargar la imagen.")
        return
    
    # Convertir la imagen a RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Calcular la imagen en escala de grises como el promedio de los canales de color
    gray_image = np.mean(image, axis=2).astype(np.uint8)
    
    # Crear un histograma para cada canal de color y para la imagen en escala de grises
    colors = ('r', 'g', 'b', 'gray')
    for i, color in enumerate(colors):
        if color == 'gray':
            hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
            plt.axvline(x=np.mean(gray_image), color='k', linestyle='dashed', linewidth=1)
        else:
            hist = cv2.calcHist([image], [i], None, [256], [0, 256])
            plt.axvline(x=np.mean(image[:, :, i]), color=color, linestyle='dashed', linewidth=1)
        
        plt.plot(hist, color=color)
    
    plt.title('Histograma')
    plt.show()

# Llamar a la función con la ruta de la imagen del panda
show_histogram('C:\\Users\\crist\\Lab1\\panda.jpg')
```

El espacio de color HSV (Hue, Saturation, Value) es un modelo de color que, al igual que el modelo de color RGB, se utiliza en el procesamiento de imágenes digitales. Este modelo representa el color en términos de matiz (Hue), saturación (Saturation) y valor (Value). A continuación se describe cada componente:

Hue (H): Representa el color en términos del tipo de color o matiz. Se mide en grados, desde 0 a 360, donde 0 o 360 es rojo, 120 es verde y 240 es azul.

Saturation (S): Es una medida de la intensidad o pureza del color (es decir, la cantidad de gris en el color). En el modelo HSV, la saturación varía de 0 (completamente desaturado, esencialmente gris) a 1 (completamente saturado, es decir, el color puro).

Value (V): También se le llama brillo, es una medida de la cantidad de luz en el color. Varía de 0 (negro puro, sin luz) a 1 (tan brillante como ese color puede ser).

El espacio de color HSV es particularmente útil porque se alinea más con la forma en que los humanos perciben el color que el modelo de color RGB. Por ejemplo, el matiz nos permite distinguir entre colores como el rojo, el verde y el azul, la saturación nos dice cuán 'intenso' es un color y el valor nos dice cuán 'brillante' es.
